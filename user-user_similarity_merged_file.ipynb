```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
data_members = pd.read_csv("members.csv", encoding = 'latin1')
data_members
```

```python
filter1 = data_members['state']=='NY'
data_members.where(filter1, inplace=True)
data_members = data_members.dropna()
data_members.head()
```

```python
data_members.shape
```

```python
data_groups = pd.read_csv("groups.csv", encoding = 'latin1')
filter1 = data_groups['state']=='NY'
data_groups.where(filter1, inplace=True)
data_groups = data_groups.dropna()
data_groups.head()
```

```python
data_groups.shape
```

```python
data_merged = pd.merge(data_members,data_groups, on='group_id')
data_merged = data_merged.dropna()
data_merged.head()
```

```python
data_merged.shape
```

```python
data_merged  = data_merged[['member_id','category_id','group_id','rating']]
data_merged.head()
```

```python
data_merged.shape
```

```python
data_merged['member_id']=data_merged['member_id'].astype(int)
data_merged['category_id']=data_merged['category_id'].astype(int)
data_merged['group_id']=data_merged['group_id'].astype(int)
data_merged = data_merged.dropna()
data_merged.head()
```
